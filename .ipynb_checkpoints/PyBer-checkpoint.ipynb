# Mergering the data frames using left join


```python
pyber_df = pd.merge(ride_df, city_df, how= "left", on=["city","city"])

pyber_df.head(10)
```

```python
pyber_df.describe()
```

# Bubble Chart
By City Type
1. Average fare for each city
2. Total number of rides for each city
3. Average number of drivers for each city


```python
#Create Urban dataframe

urban_df = pyber_df[pyber_df["type"] == "Urban"]
urban_df.head(10)
```

```python
# Create Sububan and Rural

rural_df = pyber_df[pyber_df["type"]=="Rural"]

suburban_df = pyber_df[pyber_df["type"] == "Suburban"]
```

```python
rural_df.head()
```

```python
# Calculate number of rides for Urban

urban_ride_count = urban_df.groupby(["city"]).count()["ride_id"]
urban_ride_count.head()
```

```python
# Calculate number of rides for Sububan and Rural

suburban_ride_count = suburban_df.groupby(["city"]).count()["ride_id"]

rural_ride_count = rural_df.groupby(["city"]).count()["ride_id"]
```

```python
# Calculate average of fare for All city types

urban_fare_avg = urban_df.groupby(["city"]).mean()["fare"]

suburban_fare_avg = suburban_df.groupby(["city"]).mean()["fare"]

rural_fare_avg = rural_df.groupby(["city"]).mean()["fare"]

rural_fare_avg.head()
```

```python
# Calculate average of driver_count for All city types

urban_driver_avg = urban_df.groupby(["city"]).mean()["driver_count"]

suburban_driver_avg = suburban_df.groupby(["city"]).mean()["driver_count"]

rural_driver_avg = rural_df.groupby(["city"]).mean()["driver_count"]

urban_driver_avg.head()
```

```python
# Create the Urban bubble plot

plt.scatter(urban_ride_count,urban_fare_avg,
            s=10*urban_driver_avg, 
            color="coral",
            edgecolor="k", 
            linewidths = 1,
            alpha = 0.8, 
            label = "Urban")

plt.title("PyBer Ride-Sharing Data (2019)")
plt.ylabel("Average Fare ($)")
plt.xlabel("Total Number of Rides (Per City)")
plt.grid(True)
# Add the legend.
plt.legend()
```

```python
# Create the Urban bubble plot
plt.scatter(suburban_ride_count, suburban_fare_avg,
           s=10*suburban_driver_avg,
           color = "skyblue",
           edgecolor = 'k',
           linewidths = 1,
           alpha = 0.8,
           label = "Suburban")
plt.title("PyBer Ride-Sharing Data (2019)")
plt.ylabel("Average Fare ($)")
plt.xlabel("Total Number of Rides (Per City)")
plt.grid(True)
# Add the legend.
plt.legend()
```

```python
# Create the Urban bubble plot
plt.scatter(rural_ride_count, rural_fare_avg,
           s=10*rural_driver_avg,
           color = "gold",
           edgecolor = 'k',
           linewidths = 1,
           alpha = 0.8,
           label = "Rural")
plt.title("PyBer Ride-Sharing Data (2019)")
plt.ylabel("Average Fare ($)")
plt.xlabel("Total Number of Rides (Per City)")
plt.grid(True)
# Add the legend.
plt.legend()
```

```python
plt.scatter(suburban_ride_count, suburban_fare_avg,
           s=10*suburban_driver_avg,
           color = "skyblue",
           edgecolor = 'k',
           linewidths = 1,
           alpha = 0.8,
           label = "Suburban")
plt.scatter(urban_ride_count,urban_fare_avg,
            s=10*urban_driver_avg, 
            color="coral",
            edgecolor="k", 
            linewidths = 1,
            alpha = 0.8, 
            label = "Urban")
plt.scatter(rural_ride_count, rural_fare_avg,
           s=10*rural_driver_avg,
           color = "gold",
           edgecolor = 'k',
           linewidths = 1,
           alpha = 0.8,
           label = "Rural")
```

```python
plt.subplots(figsize=(10,6))
plt.scatter(suburban_ride_count, suburban_fare_avg,
           s=10*suburban_driver_avg,
           color = "skyblue",
           edgecolor = 'k',
           linewidths = 1,
           alpha = 0.8,
           label = "Suburban")
plt.scatter(urban_ride_count,urban_fare_avg,
            s=10*urban_driver_avg, 
            color="coral",
            edgecolor="k", 
            linewidths = 1,
            alpha = 0.8, 
            label = "Urban")
plt.scatter(rural_ride_count, rural_fare_avg,
           s=10*rural_driver_avg,
           color = "gold",
           edgecolor = 'k',
           linewidths = 1,
           alpha = 0.8,
           label = "Rural")

plt.title("PyBer Ride-Sharing Data (2019)", fontsize = 20)
plt.ylabel("Average Fare ($)", fontsize = 12)
plt.xlabel("Total Number of Rides (Per City)", fontsize = 12)
plt.grid(True)

# Create a legend
lgnd = plt.legend(fontsize="12", mode="Expanded",
         scatterpoints=1, loc="best", title="City Types")
lgnd.legendHandles[0]._sizes = [75]
lgnd.legendHandles[1]._sizes = [75]
lgnd.legendHandles[2]._sizes = [75]
lgnd.get_title().set_fontsize(12)

# Incorporate a text label about circle size.
plt.text(42, 35, "Note: \nCircle size correlates \nwith driver count per city.", fontsize="12")

# Save the figure.
plt.savefig("analysis/Fig1.png")

plt.show()
```
